```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier,AdaBoostClassifier
```

```python
x,y=load_breast_cancer(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
tree= DecisionTreeClassifier(max_depth=3,random_state=23)
bagging = BaggingClassifier(base_estimator=tree, n_estimators=5, max_samples=90, bootstrap=True)

adaboost = AdaBoostClassifier(base_estimator=tree, n_estimators=5, learning_rate=0.1, random_state=23)
```

```python
tree.fit(X_train,y_train)
bagging.fit(X_train,y_train)
adaboost.fit(X_train,y_train)

```

```python
print(f"tree_train_score : {tree.score(X_train,y_train)}")
print(f"tree_test_score : {tree.score(X_test,y_test)}")
print("------------------------------")
print(f"Random_forest_bagging_train_score : {bagging.score(X_train,y_train)}")
print(f"Random_forest_bagging_test_score : {bagging.score(X_test,y_test)}")
print("------------------------------")
print(f"adaboost_train_score : {adaboost.score(X_train,y_train)}")
print(f"adaboost_test_score : {adaboost.score(X_test,y_test)}")

```
