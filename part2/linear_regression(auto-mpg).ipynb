```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python

from sklearn.metrics import fbeta_score, make_scorer,r2_score,mean_absolute_error,mean_squared_error
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']

raw_dataset = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
```

```python
raw_dataset.head()
```

```python
raw_dataset.info()
```

```python
raw_dataset.isnull().sum()
```

```python
raw_dataset.dropna(inplace=True)
```

```python
raw_dataset.isnull().sum()
```

```python
mms = MinMaxScaler()
mms.fit(raw_dataset)

raw_dataset=mms.transform(raw_dataset)
raw_dataset
df=pd.DataFrame(raw_dataset,columns=column_names)
df
```

```python
x=df.drop(columns=['MPG']).values
#x=raw_dataset.drop(columns=['MPG'])
#x.head()
```

```python
type(x)
```

```python
y=df['MPG'].values
type(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=20)
```

```python
model_r=linear_model.LinearRegression()
```

```python
model_r.fit(x_train,y_train)
```

```python
prediction=model_r.predict(x_test)
```

```python
prediction[:5]
```

```python
r2_score(prediction,y_test)
```

```python
mean_squared_error(prediction,y_test)
```
