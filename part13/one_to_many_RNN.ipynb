```python
from keras.models import Sequential
from keras.layers import*
import numpy as np
X=np.array([1,4,7,10,13,16,19,22,25,28,31,34,37,40,43])
X.shape
```

```python
y=np.array([[2,3],[5,6],[8,9],[11,12],[14,15],[17,18],[20,21],[23,24],[26,27],[29,30],[32,33],[35,36],[38,39],[41,42],[44,45]])
```

```python
y.shape
```

```python
X[4]
```

```python
y[4]
```

```python
n_time_step = 1 
n_feature =1 
```

```python
X=X.reshape((15,n_time_step,n_feature ))
X.shape
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_time_step, n_feature)))
model.add(Dense(2))
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(X, y, epochs=1000, validation_split=0.2, batch_size=3)
```

```python
y_pred=model.predict(X)
y_pred
```

```python
model.evaluate(X,y)
```
